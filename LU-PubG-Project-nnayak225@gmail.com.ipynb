```python
# 22.Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean.
# Plot it on a histogram and comment on its distribution.

mean_DamageDealt_list = []
x_values = []

def damageDealtSampleMean():
    mean_DamageDealt = round(sum(pubG['damageDealt'].sample(50))/50, 3)
    mean_DamageDealt_list.append(mean_DamageDealt)
    
for p in range(1,101):
    x_values.append(p)
    damageDealtSampleMean()

    
mean_DamageDealt_df = pd.DataFrame(mean_DamageDealt_list, columns = ['Mean Value'])
mean_DamageDealt_df["Number Of Observation"] = x_values
x_values = pd.DataFrame(x_values)
```

```python
%matplotlib inline
plt.figure(figsize=(20,8))
sns.barplot(x='Number Of Observation', y='Mean Value', data=mean_DamageDealt_df, palette="deep");
plt.xticks(rotation=70);
plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(12,7))
sns.distplot(mean_DamageDealt_list)
plt.xticks(rotation=0);
plt.show()
```

-  Most of the mean value lies between 100 to 160
